# 用来下载行情数据CSV

```python
import nest_asyncio
nest_asyncio.apply()

from ib_insync import *

import pandas as pd

contract = "TSLA"
# 初始化并连接到 TWS/Gateway
ib = IB()
ib.connect('127.0.0.1', 7496, clientId=3)
```

# 获取tsla历史分时行情

```python
# 定义合约
stock = Stock(contract, 'SMART', 'USD', primaryExchange='NASDAQ')

# 验证合约
contract_details = ib.reqContractDetails(stock)
if not contract_details:
    print("未找到合约，请检查合约定义")
    ib.disconnect()
    
bars = ib.reqHistoricalData(
    stock,
    endDateTime='',
    durationStr='1 D',
    barSizeSetting='1 min',
    whatToShow='TRADES',
    useRTH=True
)

df = pd.DataFrame(bars)
df
```

```python
if len(df) > 0:
    date = df.iloc[0]["date"]
    date = date.strftime('%Y%m%d')
    file_name = "quotes/%s_%s.csv" % (contract, date)
    df.to_csv(file_name, index=False)
```

```python
ib.disconnect()
```
