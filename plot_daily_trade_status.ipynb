```python
import yaml
import redis

def get_redis(config_file="config.yml"):
    with open(config_file, "r") as file:
        config = yaml.safe_load(file)

    redis_config = config.get("redis", {})
    return redis.Redis(**redis_config)

def build_redis_key(class_name):
    return f"{class_name}_position_manager"
```

```python
get_redis().delete(build_redis_key("StructureTradeApp"))
```

```python
import dill
data = get_redis().get(build_redis_key("StructureTradeApp"))
# data = get_redis().get(build_redis_key("TestApp"))
if data:
    data = dill.loads(data)
    positions  = data.get("positions", [])
    trade_log  = data.get("trade_log", [])
    
positions
```
